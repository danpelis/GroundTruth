```python
import pandas as pd
import os
```

```python
dir = 'submissions/'
subs = pd.DataFrame()
scores = pd.read_csv('scores.csv')
```

```python
for filename in os.listdir(dir):
    with open(os.path.join(dir, filename), 'r') as f:
        subs = pd.concat([subs, pd.read_csv(f).iloc[:, 1]], axis=1)
        print(filename)
```
