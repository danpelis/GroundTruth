```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
dir = 'submissions/'
N = 120000
scores = []
with open('scores.csv', newline='') as csvfile:
    scoreFile = csv.reader(csvfile, delimiter=',')
    next(scoreFile)
    for line in scoreFile:
        scores.append(float(line[1]))
```

```python
subs = pd.DataFrame()

for filename in os.listdir(dir):
    with open(os.path.join(dir, filename), 'r') as f:
        data = pd.read_csv(f, header=None, skiprows=1)
        subs = pd.concat([subs, data.iloc[:, 1]], axis=1)
        ids = data.iloc[:, 0]
        print(filename)
```

```python
subs
```

```python
scores
```

```python
subs = subs.replace(0, -1)
```

```python
subs_t = subs.transpose(copy=True)
subs_t
```

```python
subs.to_numpy()
subs_t.to_numpy()
product = np.dot(subs_t, subs)
product
new = np.linalg.pinv(product)
new
```

```python
subs_tx = []
for score in scores:
    subs_tx.append(N*(2*score -1))
subs_tx
```

```python
a_LS = np.dot(np.asarray(subs_tx), new)
a_LS
```

```python
subs
```

```python
subs = subs.multiply(a_LS, axis=1)
subs
```

```python
subs = subs.sum(axis=1)
subs
```

```python
s_ensem = pd.concat([ids, subs], axis=1)
s_ensem
```

```python
for index in range(0, len(s_ensem), 6):
    user_set = s_ensem.iloc[index:index+6]
    user_set.columns = ['id','score']
    user_set.sort_values(by=['score'])
    print(user_set)
    print('\n')
    
    if(index > 0):
        break
```
