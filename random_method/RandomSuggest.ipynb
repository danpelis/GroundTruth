```python
import pandas as pd
from random import shuffle
```

```python
data = 'userSuggestions.csv'
test_suggestion = pd.read_csv(data, header=None, skiprows=1)
test_suggestion.columns = ['TrackID','Predictor']
test_suggestion
```

```python
for index in range(0, len(test_suggestion), 6):
    user_set = test_suggestion.iloc[index:index+6]
    user_set.columns = ['TrackID','Predictor']
    user_set = user_set.sort_values(by=['Predictor'])
    user_set = user_set.reset_index(drop=True)
    
    preds = [0, 0, 0, 1, 1, 1]
    shuffle(preds)
    
    for row in range(6):
        if row <= 2:
            user_set['Predictor'][row] = preds[row]
        else:
            user_set['Predictor'][row] = preds[row]
    
    score_dict = user_set.set_index('TrackID').T.to_dict('list')
    
    for row in range(6):
        test_suggestion['Predictor'][index+row] = score_dict[test_suggestion['TrackID'][index+row]][0]
```
